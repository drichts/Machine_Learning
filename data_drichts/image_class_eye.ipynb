# Image classification

As someone who suffers from an eye condition that effects the macula of the eye, I am incredibly familiar with having optical coherence tomography (OCT) images taken of my eye. The macula is the most sensitive location on the retina at the back of the eye, and is responsible for color vision, central vision, and visual acuity. Disorders that damage the macula can result in anywhere from mild to complete and permanent vision loss. OCT in eye care is an imaging technique that provides micrometer resolution cross-sectional images of the retina, specifically around the macula (https://en.wikipedia.org/wiki/Optical_coherence_tomography). 


https://www.allaboutvision.com/resources/macula/

According to Swanson and Fujimoto (Swanson and Fujimoto, Biomed. Opt. Express. 2017), 30 million OCT scans are performed each year. Additionally, analyzing and interpreting the resulting images takes a signification amount of the time which could benefit from machine learning, enabling ophthalmologists (retina specialists) to quickly determine healthy from non-healthy retinas, as well as categorize various macular conditions if the images are determined to contain non-healthy areas. This could greatly accelerate the diagnosis and treatment process for patients and allow doctors to treat additional patients as well.


The data set I have chosen for image classification is a set of retinal optical coherence tomography (OCT) images. The images were selected retrospectively from patients at the Shiley Eye Institute at the University of California San Diego,  the California Retinal Research Foundation, Medical Center Ophthalmology Associates, the Shanghai First People’s Hospital, and Beijing Tongren Eye Center between July 1, 2013 and March 1, 2017. The data set consists of 84,495 images grouped into 4 disease categories. The disease categories include NORMAL, DRUSEN, CNV, and DME. DRUSEN are present in early age-related macular degeneration (AMD), CNV is chorodial neovascularization, and DME is diabetical macular edema. All are possible eye disorders that are commonly diagnosed with OCT. The images underwent multiple stages of grading to categorize and label them as well as disgarding images that were determined to have significant artifacts. The grading levels included initial quality control, followed by four ophthalmologists who graded each image independently.

The dataset can be found at this address: https://www.kaggle.com/paultimothymooney/kermany2018

As the data set is quite large, I've only included a few of images from each group.
