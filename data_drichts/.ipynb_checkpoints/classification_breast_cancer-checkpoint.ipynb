# Regression Data
The first data set I have chosen is for regression. The data set consists of intake data from a set of patients that have entered the hospital for chest pain. There is data from 297 patients, with 14 different attributes, or data such as blood pressure, age, sex, etc.
    
The data set can be found at the following address: https://www.kaggle.com/cherngs/heart-disease-cleveland-uci

The various column header abbreviations are explained below:
* age : age in years
* sex : sex of the patient (1 = male, 0 = female)
* cp : chest pain type (0: typical angina, 1: atypical angina, 2: non-anginal pain, 3: asymptomatic)
* trestbps: resting blood pressure (in mmHg)
* chol: serum cholestoral (in mg/dl)
* fbs : fasting blood sugar (>120 mg/dl = 1, <120 mg/dl = 0)
* restecg: resting electrocardiographic results (0: normal, 1: having ST-T wave abnormality, 2: showing probable or definite left ventricular hypertrophy)
* thalach: maximum heart rate achieved
* exang: exercise induced angina (1 = yes, 0 = no)
* oldpeak: ST depression induced by exercise relative to rest
* slope: the slope of the peak exercise ST segment (0: upsloping, 1: flat, 2: downsloping)
* ca: number of major vessels (0-3) colored by flouroscopy
* thal: (0: normal, 1: fixed defect, 2: reversible defect)
* condition: disease condition (0: no disease, 1: disease)

I plan on using the 'thalach' or the maximum heart rate achieved as the target, in order to see if the maximum heart rate can be predicated based on the other data.

```python
# Load the heart disease data using pandas
heart_data = pd.read_csv('heart_cleveland_upload.csv')

# Display the first 5 rows of data
heart_data.head(n=5)
```

## Display information about the data to see if cleaning is necessary

```python
heart_data.info()
```

## Look at some descriptive statistics of the data

```python
# Set pandas to output only to 2 decimal places
pd.set_option('display.float_format', lambda x: '%.2f' % x)

# Look at the statistics of the columns
heart_data.describe()
```

## Look at histograms of the data columns

```python
# Target: maximum heart rate achieved
heart_data['thalach'].hist(bins=14, figsize=(6, 4))
plt.title('Maximum heart rate',fontsize=18)  # Title
plt.xlabel("Heart rate (bpm)",fontsize=16)     # x-axis label
plt.ylabel("N",fontsize=16)       # y-axis label
plt.show()
```

```python
# All data histograms so that we can look at the distributions before and after normalization

heart_data.hist(bins=5, figsize=(15, 15))
plt.show()
```

## Set the input and target data
Target data is 'thalach' or the maximum heart rate

```python
# Set the input and target data
input_heart = heart_data[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'exang', 'oldpeak', 'slope', 'ca', 
                          'thal', 'condition']].values
target_heart = heart_data[['thalach']].values

# Print the shapes of the data to make sure the size is the same (number of rows)
print(f'Input shape: {np.shape(input_heart)}')
print(f'Target shape: {np.shape(target_heart)}')
```

## Normalize the data and make sure the distribution is the same as above

```python
# Apply a standard normalization to the input data
stnd_scaler = StandardScaler().fit(input_heart)
stnd_scaler.fit(input_heart)
input_heart_stnd = stnd_scaler.transform(input_heart)

# Apply a standard normalization to the target data
stnd_scaler_tar = StandardScaler().fit(target_heart)
stnd_scaler_tar.fit(target_heart)
target_heart_stnd = stnd_scaler_tar.transform(target_heart)

# Show the target distribution
plt.hist(target_heart_stnd, 14)
plt.title('Maximum heart rate after normalization',fontsize=18)  # Title
plt.ylabel("N",fontsize=16)  # y-axis label
plt.show()
```

```python
# Show the input distribution
num_plots = 13  # The number of input columns

# The titles for each of the subplots
titles = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'condition']

# Create the figure
fig, axes = plt.subplots(4, 4, figsize=(16, 16))
for i, ax in enumerate(axes.flatten()):
    if i == 13:
        break  # Only 13 plots to make, don't need the last 3
    ax.hist(input_heart_stnd[:, i], 4)  # Plot the histogram
    ax.set_title(titles[i], fontsize=15)  # Set the title for each subplot
plt.show()
```

## Now, the target is binary, but set to 'B' or 'M', need to set it to 0 or 1.
## B = 0
## M = 1

```python
# Check the number of M and B
print(f'Number of M before: {len(np.squeeze(np.argwhere(target_can == "M")))}')
print(f'Number of B before: {len(np.squeeze(np.argwhere(target_can == "B")))}')
print()

# Set B to 0
target_can[target_can == 'B'] = 0

# Set M to 1
target_can[target_can == 'M'] = 1
print(f'Target shape: {np.shape(target_can)}')

# Check it's the same number afterwards
print(f'Number of M after: {len(np.squeeze(np.argwhere(target_can == 1)))}')
print(f'Number of B after: {len(np.squeeze(np.argwhere(target_can == 0)))}')
```

```python
temp_titles = ['Mean radius', 'Std. error of the radius', 'Worst radius']
fig, axes = plt.subplots(3, 1, figsize=(8, 20))
for i, ax in enumerate(axes.flatten()):
    ax.hist(input_can[:, i*10], 15)  # Plot the histogram
    ax.set_title(temp_titles[i], fontsize=24)# Set the title for each subplot
    ax.set_xlabel('Radius', fontsize=20)
    ax.set_ylabel('N', fontsize=20)
    ax.tick_params(labelsize=18)

plt.subplots_adjust(hspace=0.3)
plt.show()
fig.savefig('Radius.png', dpi=500)
```

# Image classification
The data set I have chosen for image classification is a set of retinal optical coherence tomography (OCT) images. The data set consists of 84,495 images grouped into 4 disease categories. The disease categories include NORMAL, DRUSEN, CNV, and DME. DRUSEN are present in early age-related macular degeneration (AMD), CNV is chorodial neovascularization, and DME is diabetical macular edema. All are possible eye disorders that are commonly diagnosed with OCT. The analysis of the images can be time consuming so machine learning could applied to the images. The dataset is already organized into 3 smaller sets for training, testing, and validation, each with sets of the 4 disease categories within it.

The dataset can be found at this address: https://www.kaggle.com/paultimothymooney/kermany2018

As the data set is quite large, I've only included a couple of images from each group.

Below is an image demonstrating the differences between the different disease types.


Figure 2. Representative Optical Coherence Tomography Images and the Workflow Diagram [Kermany et. al. 2018] http://www.cell.com/cell/fulltext/S0092-8674(18)30154-5

(A) (Far left) choroidal neovascularization (CNV) with neovascular membrane (white arrowheads) and associated subretinal fluid (arrows). (Middle left) Diabetic macular edema (DME) with retinal-thickening-associated intraretinal fluid (arrows). (Middle right) Multiple drusen (arrowheads) present in early AMD. (Far right) Normal retina with preserved foveal contour and absence of any retinal fluid/edema.
