# photometric redshift with Machine Learning 
This project aims to estimate the photometric redshift of distance galaxies; z. Then a redshift of a galaxy (a real number) can be used to find a galaxy's distance. 







# The following text from Wikipedia (optional, it could be only a link):


A photometric redshift is an estimate for the recession velocity of an astronomical object 
such as a galaxy or quasar, made without measuring its spectrum. The technique uses photometry 
(that is, the brightness of the object viewed through various standard filters, each of which 
lets through a relatively broad passband of colours, such as red light, green light, or blue light) 
to determine the redshift, and hence, through Hubble's law, the distance, of the observed object.
The technique was developed in the 1960s, but was largely replaced in the 1970s and 1980s by 
spectroscopic redshifts, using spectroscopy to observe the frequency (or wavelength) of 
characteristic spectral lines, and measure the shift of these lines from their laboratory 
positions. The photometric redshift technique has come back into mainstream use since 2000, 
as a result of large sky surveys conducted in the late 1990s and 2000s which have detected a 
large number of faint high-redshift objects, and telescope time limitations mean that only a 
small fraction of these can be observed by spectroscopy. Photometric redshifts were originally 
determined by calculating the expected observed data from a known emission spectrum at a range of 
redshifts. The technique relies upon the spectrum of radiation being emitted by the object having 
strong features that can be detected by the relatively crude filters.
As photometric filters are sensitive to a range of wavelengths, and the technique relies on making 
many assumptions about the nature of the spectrum at the light-source, errors for these sorts of 
measurements can range up to z = 0.5, and are much less reliable than spectroscopic determinations.
In the absence of sufficient telescope time to determine a spectroscopic redshift for each object, the technique of photometric redshifts provides a method to determine an at least qualitative characterization of a redshift. For example, if a Sun-like spectrum had a redshift of z = 1, it would be brightest in the infrared rather than at the yellow-green color associated with the peak of its blackbody spectrum, and the light intensity will be reduced in the filter by a factor of two (i.e. 1+z) (see K correction for more details on the photometric consequences of redshift).[3]
Other means of estimating the redshift based on alternative observed quantities have been developed,
like for instance morphological redshifts applied to galaxy clusters which rely on geometric
measurements [4] In recent years, Bayesian statistical methods and artificial neural networks
have been used to estimate redshifts from photometric data. Wiki


# We need NumPy and mathplotlib to arrange the input data and for plotting.
# PolynomialFeatures:
Generate a new feature matrix consisting of all polynomial combinations of the features
with degree less than or equal to the specified degree. For example, if an input sample is 
two-dimensional and of the form [a, b], the degree-2 polynomial features are 
[1, a, b, a^2, ab, b^2]. For more information, see the following page:
    
https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html    

import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import PolynomialFeatures



The data used in this notebook is for estimating photometric redshift using information from galaxy fluxes in different filters. The data contains 7 columns, and the data size is 713. Each column is the difference between two selected filters, such as i-z. The information in the table will be used to predict 713 corresponding targets (redshifts). The data is taken from the catalogue from the following paper (Santini et al., 2019). We can use different models for z predictions.



```python
#Load the data sets
inp= np.load('inp_redshift.npy')
tar= np.load('tar_redshift.npy')


print('Input size/dim = ', np.shape(inp))   # Checking the size and dimension of the data (input)
print('target size/dim = ', np.shape(tar))  # Checking the size and dimension of the data (target)
```

```python
dim = np.shape(inp)[1]  # Assigning the dimension of the input (to dim)
print (dim)
```

# Exploring the data and seeing the distribution of all 7 columns

```python
for k1 in range(dim):  # ploting the histogram of the 7 columns
    plt.figure(k1)
    plt.hist(inp[:,k1])
    plt.title('Column- '+str(k1))
    plt.ylabel('N')
    plt.xlabel("X"+str(k1))
        
```

# Randomly separate 713 samples to the training set (75%) and validation set (25%). The corresponding targets can be separated too. 

For more information see:

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html


```python
from sklearn.model_selection import train_test_split
inp_tr,inp_va,tar_tr, tar_va = train_test_split(inp,tar ,test_size=0.25)

print ('training set == ',np.shape(inp_tr),np.shape(tar_tr),',, validation set == ', np.shape(inp_va),np.shape(tar_va))
```

# Normalization.

Line #1: Import a model, for normalization, like StandardScaler (https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html)

Line #2: fitting (finding the parameters of the model based on the training set). Assign a name (scaler_S which can transform a new data after training)

Line #3: Predicted (transformed) values for the training set

Line #4: Predicted (transformed) values for the validation set (using the model that has been trained using the training set).



```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler  #line #1

scaler_S= StandardScaler().fit(inp_tr)  # line #2
inp_tr_Nstd= scaler_S.transform(inp_tr) # line # 3


inp_va_Nstd= scaler_S.transform(inp_va)  # Line #4
```

# See the normalized results for the validation set (the same can be done for the training set, and comparisons may be made to see the differences, for example)

```python

for k1 in range(dim):
    plt.figure(k1)
    plt.hist(inp_va_Nstd[:,k1])
    plt.title('Column- '+str(k1) + "  (Normalized STD)")
    plt.ylabel('N')
    plt.xlabel("X"+str(k1))
```

# Feature extraction

If degree=1, you will not have any feature extraction.

For example, if degree=2, for two columns a and b,  you will have the following combination of features
: [1, a, b, a^2, ab, b^2]

If Degree = 2, for 7 columns, you will have a matrix with 36 columns ( 35 combinations + one bias).

If Degree = 6, for example, the engineered input could be very complicated (high dimensionality = 1715+1, which is higher than the size of the data and is not good generally)


```python
degree= 4


# for example, if  degree = 2 and if we have 7 columns, then the following line creates a table of 36 
#columns of the possible combination to power 2combinations
poly = PolynomialFeatures(degree,include_bias=True)


poly.fit(inp_tr_Nstd)  # fitting the model based on the training set

inp_tr_Nstd_Ext=poly.transform(inp_tr_Nstd)  # transforming the training set 

inp_va_Nstd_Ext=poly.transform(inp_va_Nstd)  # transforming the validation set 





# making a one coulumn target sets  (mx1 : a columns vector for target)

tar_tr = np.reshape(tar_tr,(-1,1))
tar_va = np.reshape(tar_va,(-1,1))


# cheking the dimension of the engineered inputs

print ("traing (input) size/dim == ", np.shape(inp_tr_Nstd_Ext))
print ("traing (target) size/dim == ", np.shape(tar_tr))


print ("vallidation (input) size/dim == ", np.shape(inp_va_Nstd_Ext))
print ("vallidation (target) size/dim == ", np.shape(tar_va))



```

# Regularization L2: Adding a penalty term to the error function to keep the weights (W) small to avoid overfitting 



```python
# new dimentoin of engineered data

dim_lamb=np.shape(inp_tr_Nstd_Ext)[1]


#The value of the penalty, play with it  (the lower, more overfitting with strong model. with strong models use hight values for lamb)

lamb=.0003

L2=lamb *np.identity(dim_lamb)

```

# Training the model (the closed-form) and the predicting (yh_tr: predictions for training and validation sets)

```python
#The first three lines are related to training the model (finding weights:W)

inv_XTX= np.linalg.inv(inp_tr_Nstd_Ext.T.dot(inp_tr_Nstd_Ext)+L2)
upper = (inp_tr_Nstd_Ext.T).dot(tar_tr)
theta= inv_XTX.dot(upper)


# predictions

yh_tr= inp_tr_Nstd_Ext.dot(theta)
yh_va= inp_va_Nstd_Ext.dot(theta)
                        

```

# This is a method for plotting training and validation sets side by side and comparing the statistics.
# The aim is to see more similar behaviour. If overfitting, play with the lambda: the hyperparameter for controlling overfitting.

# You might use another method; just make sure that training and validation sets are compared visually and using statistics.


```python



PY_tr=(yh_tr)
PY_va=(yh_va)

Y_tr = tar_tr
Y_va = tar_va



def mu_med_sig(pre,tar):
    mu = np.mean(pre-tar)
    med = np.median(pre-tar)
    std = np.std(pre-tar)
    return mu,med,std


fig,axarr = plt.subplots(1,2,figsize=(11,5))



axarr[0].scatter(Y_tr,PY_tr,edgecolor='black',linewidths=2,facecolor=None)
axarr[0].scatter(Y_tr,PY_tr,facecolor='C0',label='Training set')
axarr[0].set_xlabel('Target Redshift, $z_{\mathrm{tar}}$')
axarr[0].set_ylabel('Predicted Redsift')
axarr[0].legend(loc=4)
mu,med,sig = mu_med_sig(PY_tr,Y_tr)
axarr[0].text(0.05,0.95,'$\mu$: %0.4f \nmedian: %0.4f \n$\sigma$: %0.4f'%(mu,med,sig),
              ha='left',va='top',transform=axarr[0].transAxes)

axarr[1].scatter(Y_va,PY_va,edgecolor='black',linewidths=2,facecolor=None)
axarr[1].scatter(Y_va,PY_va,facecolor='C1',label='Validation set')
axarr[1].set_xlabel('Target Redshift, $z_{\mathrm{tar}}$')
axarr[1].set_ylabel('Predicted Redsift')
axarr[1].legend(loc=4)
mu,med,sig = mu_med_sig(PY_va,Y_va)
axarr[1].text(0.05,0.95,'$\mu$: %0.4f \nmedian: %0.4f \n$\sigma$: %0.4f'%(mu,med,sig),
              ha='left',va='top',transform=axarr[1].transAxes)

for ax in axarr:
    xmin,xmax=-0.1,1.7
    ax.set_xlim(xmin,xmax)
    ax.set_ylim(xmin,xmax)
    ax.plot([xmin,xmax],[xmin,xmax],color='black',ls='dashed',zorder=-1)
    
#fig.savefig('regression_colours.pdf',bbox_inches='tight')
```
